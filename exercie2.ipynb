# POLOGNE : Carte du déplacement de frontière de 1945

Après la Ie Guerre Mondiale, le Traité de Versailles recrée une Pologne indépendante; les polonais reconquièrent des territoires à l'Est par une guerre avec leurs voisins et surtout l'Union Soviétique (miracle de la Vistule, 1920).

Le 1er septembre 1939, Le cuirassé Schleswig-Holstein ouvre le feu sur la Westerplatte à Dantzig; les troupes allemandes franchissent la frontière, déclenchant la 2e Guerre Mondiale. La Pologne est réduite au "Gouvernement Général" compris entre Lublin et Varsovie ; les régions occidentales (Posnanie, Wroclaw-Breslau) sont annexées à l'Allemagne et les populations remplacées par des immigrés d'origine germanique venant d'autres pays européens. En 1945-1946, la frontière polonaise est repoussée à l'Ouest jusqu'à l'Oder par les accords de Yalta et Potsdam, englobant des territoires qui n'avaient jamais été polonais, la Prusse Occidentale est attribuée à la Pologne, pendant que les territoires de l'Est (Wilno, Brest-Litovsk, Lvov) sont annexés à l'Union Soviétique donnant lieu à la migration de plusieurs millions de personnes qui vont remplacer les Allemands expulsés.

Le souvenir douloureux des événements de la seconde guerre mondiale est très présent dans la mémoire nationale. La Pologne perd 6 millions d'habitants sur une population de 30, dont une moitié de victimes Juives.
A Gdansk en 1970, 1980 puis au cours de la décennie 1980, les mouvements des ouvriers des chantiers navals participent à la chute du Rideau de Fer.

```python
#Création de la carte
Polo = folium.Map([52.237049, 21.017532], zoom_start = 5, tiles = 'openstreetmap')
```

```python
# afficher plusieur fonds de carte 
tile = folium.TileLayer('Stamen Terrain').add_to(Polo)
tile = folium.TileLayer('Stamen Toner').add_to(Polo)
tile = folium.TileLayer('stamenwatercolor').add_to(Polo)
tile = folium.TileLayer('cartodbpositron').add_to(Polo)
tile = folium.TileLayer('cartodbdark_matter').add_to(Polo)
```

```python
#ajout des fichiers geojson à la carte
folium.GeoJson(Pologne_2020, name='la pologne de 2020').add_to(Polo)
folium.GeoJson(Pologne_1945, name='la pologne de 1945',show=False).add_to(Polo)
folium.GeoJson(Pologne_1925, name='la pologne de 1925',show=False).add_to(Polo)
folium.GeoJson(Pologne_1772, name='la pologne de 1772',show=False).add_to(Polo)
```
